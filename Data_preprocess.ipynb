```python
#importing datasets  
data_set= pd.read_csv('Air_quality.csv')
```

```python
# Drop date
data_set.drop('Data',
  axis='columns', inplace=True)
```

```python
# drop the column where time is missing
data_set.dropna(subset = ["Time"], inplace=True)
```

```python
# Show the data after drop
data_set
```

```python
#Set the time as a index column
data_set.set_index(['Time'],inplace=True)
```

```python
#Show data
data_set
```

```python
# handling missing data(Replacing missing data with the mean value) 
data_set.fillna(data_set.mean())
```

```python
#Extracting Independent Variable  
x= data_set.iloc[:, data_set.columns != 'PM2.5'].values  
```

```python
# show the independent variable
x
```

```python
#Extracting Dependent variable  
y= data_set.iloc[:, 3].values 
```

```python
# Showing the dependent variable
y
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=0)  
```

```python
#Feature Scaling of datasets  
from sklearn.preprocessing import StandardScaler  
st_x= StandardScaler()  
x_train= st_x.fit_transform(x_train)  
x_test= st_x.transform(x_test)  
```
